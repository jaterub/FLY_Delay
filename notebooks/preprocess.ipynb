```python
def onehot_encode(df, column_dict):
    df = df.copy()
    for column, prefix in column_dict.items():
        dummies = pd.get_dummies(df[column], prefix=prefix)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
```

```python
df = onehot_encode(df,column_dict={'AIRLINE': 'AL','ORIGIN_AIRPORT': 'OA','DESTINATION_AIRPORT': 'DA'})
```

```python

def map_unique_values(df, feature):
    unique_values = df[feature].unique()
    mapping_dict = {value: code for code, value in enumerate(unique_values, 1)}
    df[f'{feature}_CODE'] = df[feature].map(mapping_dict)


```
